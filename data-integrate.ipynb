```python
def format_excel_serial_time(serial_time):
    """
    Convert an Excel serial time (fraction of a day) to a 12-hour format string.
    
    For example, 0.812500 corresponds to 7:30 PM.
    """
    # There are 86400 seconds in a day.
    total_seconds = serial_time * 86400
    hours = int(total_seconds // 3600)
    minutes = int((total_seconds % 3600) // 60)
    
    # Determine period (AM or PM) and adjust hour to 12-hour format.
    period = "AM" if hours < 12 else "PM"
    
    display_hour = hours % 12
    if display_hour == 0:
        display_hour = 12

    return f"{display_hour}:{minutes:02d} {period}"

```

```python
def format_excel_serial_date(serial_date, format_str="%m/%d/%Y"):
    """
    Convert an Excel serial date to a formatted date string.
    
    Parameters:
        serial_date (int or float): The Excel serial date number (e.g., 45658).
        format_str (str): The output format string (default: "%m/%d/%Y").
                          You can adjust it to your preferred format.
    
    Returns:
        str: The formatted date string.
    """
    # Excel serial dates are counted from '1899-12-30'
    date_value = pd.to_datetime(serial_date, origin="1899-12-30", unit="D")
    return date_value.strftime(format_str)
```

```python
def convert_time(time_str):
    """
    Convert a time string in 24-hour format (e.g., "24:00", "14:30") 
    to a 12-hour AM/PM formatted string.
    If the input is "24:00", it will be treated as midnight ("12:00 AM").
    """
    time_str = str(time_str)
    # print(type(time_str))

    time_str = time_str.strip()  # remove any leading/trailing spaces
    # If the time is "24:00", replace it with "00:00" so it can be parsed correctly.
    if time_str == "24:00":
        time_str = "00:00"
    
    try:
        # Parse the time from the 24-hour format
        formatted_time =  datetime.strptime(time_str, "%H:%M").strftime("%I:%M %p")
    except ValueError as e:
        # In case of a format error, just return the original string
        print(e)
        formatted_time = time_str
    return formatted_time
```

```python
# mappingroad = {
#     "ทางตรง+ไม่มีความลาดชัน": "straight road",
#     "ทางโค้งกว้าง+ไม่มีความลาดชัน": "wide curve",
#     "ทางสี่แยก": "four‑way intersection",
#     "ทางตรง+ที่ลาดชัน": "sharp curve",
#     "ทางโค้งกว้าง+ที่ลาดชัน": "roundabout",
#     "ทางสามแยก(T)": "t‑intersection",
#     "ทางเชื่อมเข้าพื้นที่สาธารณะ/เชิงพาณิชย์": "connecting to public/commercial area",
#     "ทางตรง": "straight road",
#     "ทางโค้งหักศอก+ไม่มีความลาดชัน": "sharp curve",
#     "ทางเชื่อมเข้าพื้นที่ส่วนบุคคล": "connecting to private area",
#     "ทางสามแยก(Y)": "y‑intersection",
#     "ทางโค้งกว้าง": "wide curve",
#     "ทางแยกต่างระดับ/Ramps": "grade‑separated intersection/ramps",
#     "ทางโค้งหักศอก+ที่ลาดชัน": "roundabout",
#     "อื่นๆ": "other",
#     "ทางร่วม": "merge lane",
#     "ทางเชื่อมเข้าบริเวณหน้าโรงเรียน": "connecting to school area",
#     "ทางคนเดินเท้า": "pedestrian path",
# }

# def translate(x):
#     x = x.strip()
#     return mappingroad.get(x, pd.NA)


def translate_feature(feature: str) -> str:
    """
    Translates a Thai road feature description into an English category.

    Mapping rules (in order of priority):
      - Grade-separated intersection/ramps: if "แยกต่างระดับ" or "Ramps" is present.
      - Four-way intersection: if "สี่แยก" is present.
      - T-/Y-intersections: if "สามแยก" or "แยกรูปตัว" is present.
          • If the feature contains "T", returns "t-intersection".
          • If the feature contains "Y", returns "y-intersection".
          • Otherwise returns "other".
      - Zebra crossing: if "ม้าลาย" is present.
      - Pedestrian path: if "คนเดินเท้า" is present.
      - Connecting to school area: if "สถานศึกษา", "สถานที่ราชการ", or "โรงเรียน" is present.
      - Connecting to private area: if "ส่วนบุคคล" is present.
      - Connecting to public/commercial area: if "สาธารณะ" or "พาณิชย์" is present.
      - Merge lane: if "ร่วม" is present.
      - U-turn area: if "กลับรถ" is present.
      - Roundabout: if "วงเวียน" is present.
      - Motorcycle lane: if "จักรยาน" is present.
      - Bridge: if "สะพาน" is present.
      - Lane-changing area: if "เปลี่ยนความกว้าง" is present.
      - Other (e.g., rail crossing or underpass): if "รถไฟ" or "ลอด" is present.
      - Sharp curve: if "โค้งหักศอก" is present.
      - Wide curve: if "โค้งกว้าง" or "โค้งปกติ" is present.
      - Straight road: if "ตรง" is present.
      - Otherwise, returns "other".
    
    If the input is missing or 'nan', it will return "other".
    """
    feature = str(feature)
    if feature is None or feature.strip() == "" or feature.lower() == "nan":
        return "other"

    # Use the original text for pattern matching
    f = feature

    # Intersection and ramp types
    if "แยกต่างระดับ" in f or "Ramps" in f:
        return "grade-separated intersection/ramps"
    if "สี่แยก" in f:
        return "four-way intersection"
    if "สามแยก" in f or "แยกรูปตัว" in f:
        if "T" in f:
            return "t-intersection"
        elif "Y" in f:
            return "y-intersection"
        else:
            return "other"

    # Pedestrian related
    if "ม้าลาย" in f:
        return "zebra crossing/pedestrian crossing"
    if "คนเดินเท้า" in f:
        return "pedestrian path"

    # Connecting to specific areas
    if "สถานศึกษา" in f or "สถานที่ราชการ" in f or "โรงเรียน" in f:
        return "connecting to school area"
    if "ส่วนบุคคล" in f:
        return "connecting to private area"
    if "สาธารณะ" in f or "พาณิชย์" in f:
        return "connecting to public/commercial area"

    # Other specific features
    if "ร่วม" in f:
        return "merge lane"
    if "กลับรถ" in f:
        return "u-turn area"
    if "วงเวียน" in f:
        return "roundabout"
    if "จักรยาน" in f:
        return "motorcycle lane"
    if "สะพาน" in f:
        return "bridge (across river/canal)"
    if "เปลี่ยนความกว้าง" in f:
        return "lane-changing area"
    if "รถไฟ" in f or "ลอด" in f:
        return "other"

    # Curve and straight road types
    if "โค้งหักศอก" in f:
        return "sharp curve"
    if "โค้งกว้าง" in f or "โค้งปกติ" in f:
        return "wide curve"
    if "ตรง" in f:
        return "straight road"

    return "other"


# def translate_features(features: list) -> list:
#     """
#     Translates a list of Thai road feature descriptions into their English categories.
#     """
#     return [translate_feature(feat) for feat in features]
```

```python
def convert_weather(value):
    if value == "แจ่มใส":
        return "clear"
    elif value == "มีหมอก/ควัน/ฝุ่น":
        return "foggy"
    elif value == "ฝนตก":
        return "rainy"
    elif value == "มืดครึ้ม":
        return "dark"
    elif value == "อื่นๆ":
        return "others"
    elif value == "ฝนตก":
        return "rainy"
    elif value == "ภัยธรรมชาติ เช่น พายุ น้ำท่วม":
        return "natural disaster" 
```

```python
def create_custom_category(d: str) -> str:
    """
    Fallback logic for descriptions that don't match any primary rule.
    Based on common keywords, this function tries to generate a category.
    """
    # Check for rollover/fallen events by looking for keywords
    if ("เสียหลัก" in d or "ตกถนน" in d):
        if "โค้ง" in d:
            return "rollover/fallen on curved road"
        else:
            return "rollover/fallen on straight road"
    # If the description involves collision events, decide based on keywords
    if "ชน" in d:
        # Rear-end collision detection
        if "ท้าย" in d:
            return "rear-end collision"
        # Check for turning or retreating collision
        if "เลี้ยว" in d or "ถอย" in d:
            return "turning/retreating collision"
        # Check for side collision indications
        if "ด้านข้าง" in d or "เฉี่ยว" in d:
            return "side collision"
        # Check for head-on collision cues
        if "ตรงกันข้าม" in d:
            return "head-on collision (not overtaking)"
        # Check if it might be a collision with an obstruction
        if "ป้าย" in d or "วัสดุ" in d or "สิ่งกีดขวาง" in d:
            return "collision with obstruction (on road surface)"
    # If no custom rule applies, return "other"
    return "other"


def translate_accident(description: str) -> str:
    """
    Translates a Thai accident description into one of the following English categories:
    
    - 'other'
    - 'rollover/fallen on straight road'
    - 'head-on collision (not overtaking)'
    - 'collision at intersection corner'
    - 'collision with obstruction (on road surface)'
    - 'rear-end collision'
    - 'pedestrian collision'
    - 'rollover/fallen on curved road'
    - 'collision during overtaking'
    - 'turning/retreating collision'
    - 'side collision'
    
    The mapping is based on keyword matching. If no primary rule applies, a custom
    fallback function is used.
    """
    description = str(description)
    # Return "other" for missing details.
    if description is None or description.strip() == "" or description.lower() in ["nan", "ไม่มีรายละเอียด"]:
        return "other"
    
    d = description.strip()

    # Predefined mapping rules
    if "ชนในทิศทางตรงกันข้าม (ไม่ใช่การแซง)" in d:
        return "head-on collision (not overtaking)"
    if "ชนเป็นมุมบริเวณทางแยก" in d:
        return "collision at intersection corner"
    if "ชนสิ่งกีดขวาง" in d:
        return "collision with obstruction (on road surface)"
    if "ชนคนเดินเท้า" in d:
        return "pedestrian collision"
    if "ชนท้าย" in d:
        return "rear-end collision"
    if ("พลิกคว่ำ" in d or "ตกถนนในทางตรง" in d) and "โค้ง" not in d:
        return "rollover/fallen on straight road"
    if ("พลิกคว่ำ" in d or "ตกถนน" in d) and "โค้ง" in d:
        return "rollover/fallen on curved road"
    if "แซง" in d:
        return "collision during overtaking"
    if "เลี้ยว/ถอยชน" in d:
        return "turning/retreating collision"
    if "ชนด้านข้าง" in d:
        return "side collision"

    # Fallback: create your own category for unmatched descriptions
    return create_custom_category(d)
```

```python
def translate_vehicle(vehicle: str) -> str:
    """
    Translates a Thai vehicle type into its corresponding English category.
    
    The mapping is defined as follows:
      - If the text contains "รถจักรยานยนต์", returns "motorcycle".
      - If the text contains "รถยนต์นั่งส่วนบุคคล" or "รถยนต์นั่งสาธารณะ", returns "private/passenger car".
      - If the text contains "รถปิคอัพบรรทุก 4 ล้อ", returns "4-wheel pickup truck".
      - If the text is exactly "อื่นๆ", returns "other".
      - If the text contains "รถสามล้อเครื่อง", returns "motorized tricycle".
      - If the text contains "รถบรรทุกมากกว่า 10 ล้อ" or "รถพ่วง", returns "large truck with trailer".
      - If the text contains "รถตู้", returns "van".
      - If the text contains "รถบรรทุก 6 ล้อ", returns "6-wheel truck".
      - If the text contains "รถบรรทุกมากกว่า 6 ล้อ" and "ไม่เกิน 10 ล้อ", returns "7-10-wheel truck".
      - If the text contains "รถจักรยาน", returns "bicycle".
      - If the text contains "รถปิคอัพโดยสาร", returns "passenger pickup truck".
      - If the text contains "คนเดินเท้า", returns "pedestrian".
      - If the text contains "รถโดยสารขนาดใหญ่", returns "large passenger vehicle".
      - If the text contains "รถอีแต๋น" or "เพื่อการเกษตร", returns "tractor/agricultural vehicle".
      - If the text contains "รถสามล้อ" (and did not match "รถสามล้อเครื่อง"), returns "three-wheeled vehicle".
      - Otherwise, returns "other".
    """
    vehicle = str(vehicle)
    if vehicle is None or vehicle.strip() == "" or vehicle.lower() == "nan":
        return "other"
    
    v = vehicle.strip()
    
    # Order of checks is important to resolve potential ambiguities.
    if "รถจักรยานยนต์" in v:
        return "motorcycle"
    if "รถยนต์นั่งส่วนบุคคล" in v or "รถยนต์นั่งสาธารณะ" in v:
        return "private/passenger car"
    if "รถปิคอัพบรรทุก 4 ล้อ" in v:
        return "4-wheel pickup truck"
    if v == "อื่นๆ":
        return "other"
    if "รถสามล้อเครื่อง" in v:
        return "motorized tricycle"
    if "รถบรรทุกมากกว่า 10 ล้อ" in v or "รถพ่วง" in v:
        return "large truck with trailer"
    if "รถตู้" in v:
        return "van"
    if "รถบรรทุก 6 ล้อ" in v:
        return "6-wheel truck"
    if "รถบรรทุกมากกว่า 6 ล้อ" in v and "ไม่เกิน 10 ล้อ" in v:
        return "7-10-wheel truck"
    if "รถจักรยาน" in v:
        return "bicycle"
    if "รถปิคอัพโดยสาร" in v:
        return "passenger pickup truck"
    if "คนเดินเท้า" in v:
        return "pedestrian"
    if "รถโดยสารขนาดใหญ่" in v:
        return "large passenger vehicle"
    if "รถอีแต๋น" in v or "เพื่อการเกษตร" in v:
        return "tractor/agricultural vehicle"
    if "รถสามล้อ" in v:
        # This check comes after checking for "รถสามล้อเครื่อง"
        return "three-wheeled vehicle"
    
    return "other"
```

```python
def integrate_datetime_columns(df):
    """
    Integrates separate date and time columns into two new columns:
      - 'incident_datetime' from 'วันที่เกิดเหตุ' and 'เวลา'
      - 'report_datetime' from 'วันที่รายงาน' and 'เวลาที่รายงาน'
    
    The output format is: M/D/YYYY H:MM:SS AM/PM.
    
    It also handles cases where the time value is given as a fraction of a day.
    """
    # Define output formats for date and time based on OS
    if platform.system() == "Windows":
        full_format = '%#m/%#d/%Y %#I:%M:%S %p'
        time_format = '%#I:%M:%S %p'
    else:
        full_format = '%-m/%-d/%Y %-I:%M:%S %p'
        time_format = '%-I:%M:%S %p'
    
    def convert_time(time_str):
        """
        Convert a time string to a formatted time.
        If the string contains a colon, assume it's already in a conventional time format.
        Otherwise, treat it as a fraction of a day.
        """
        ts = str(time_str).strip()
        if ":" in ts:
            # Try converting the conventional time string
            try:
                t = pd.to_datetime(ts).time()
                return t.strftime(time_format)
            except Exception:
                return ts  # fallback to the original string if conversion fails
        else:
            try:
                # Treat ts as a fraction of a day
                fraction = float(ts)
                total_seconds = fraction * 86400  # 24*60*60
                # Create a time object starting from midnight
                t = (datetime.datetime(1900, 1, 1) + datetime.timedelta(seconds=total_seconds)).time()
                return t.strftime(time_format)
            except Exception:
                return ts

    def combine_datetime(date_str, time_str):
        """
        Combines a date string and a time string into a single datetime string in the desired format.
        """
        try:
            # Convert the date string to a datetime object
            date_obj = pd.to_datetime(date_str)
            # Format the date part as MM/DD/YYYY (zero-padded)
            date_formatted = date_obj.strftime("%m/%d/%Y")
        except Exception:
            return None  # or you can return date_str + " " + time_str as fallback

        # Convert time_str using the helper function
        time_formatted = convert_time(time_str)
        # Combine date and time strings
        combined_str = f"{date_formatted} {time_formatted}"
        try:
            # Convert back to datetime and then format it using full_format
            combined_dt = pd.to_datetime(combined_str)
            return combined_dt.strftime(full_format)
        except Exception:
            return combined_str  # fallback if conversion fails

    # Create the new datetime columns using a row-wise apply
    df['incident_datetime'] = df.apply(lambda row: combine_datetime(row['วันที่เกิดเหตุ'], row['เวลา']), axis=1)
    df['report_datetime'] = df.apply(lambda row: combine_datetime(row['วันที่รายงาน'], row['เวลาที่รายงาน']), axis=1)
    return df
```

```python
def combine_with_base(base_df, *dfs):
    """
    Combine multiple DataFrames using base_df as the target schema.
    Each of the other DataFrames is reindexed to match the base_df's columns.
    
    Parameters:
        base_df (DataFrame): The base DataFrame whose columns define the final schema.
        *dfs (DataFrame): One or more DataFrames to combine with the base.
    
    Returns:
        DataFrame: The combined DataFrame with columns matching base_df.
    """
    # Use the base DataFrame's columns as the desired schema.
    base_cols = list(base_df.columns)
    
    # Create a list starting with a copy of the base DataFrame.
    aligned_dfs = [base_df.copy()]
    
    # Reindex each provided DataFrame to match the base columns.
    for df in dfs:
        aligned_df = df.reindex(columns=base_cols)
        aligned_dfs.append(aligned_df)
    
    # Concatenate all DataFrames row-wise.
    combined_df = pd.concat(aligned_dfs, ignore_index=True, sort=False)
    
    return combined_df
```
